```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('MP_Yearly_Consolidated_Rainfall_Data_1_1.csv')
print(data.isnull().sum())
```

```python
data.columns = [col.replace(' ', '_').replace('.', '').replace('(', '').replace(')', '') for col in data.columns]
```

```python
rainfall_cols = ['District', 'Long_Term_Normal_Rainfall_LTNR', '2011_Actual_Rainfall', '2012_Actual_Rainfall', '2013_Actual_Rainfall', '2014_Actual_Rainfall']

df = data[rainfall_cols]
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['Long_Term_Normal_Rainfall_LTNR'], df['2013_Actual_Rainfall'], color='blue', alpha=0.7)
plt.title('Scatter Plot of LTNR vs Actual Rainfall (2013)')
plt.xlabel('Long Term Normal Rainfall (mm)')
plt.ylabel('2013 Actual Rainfall (mm)')
plt.grid(True)
plt.show()
```

```python
rainfall_data = df[['2011_Actual_Rainfall', '2012_Actual_Rainfall', '2013_Actual_Rainfall', '2014_Actual_Rainfall']]
plt.figure(figsize=(10,6))
plt.boxplot([rainfall_data[col] for col in rainfall_data.columns], labels=rainfall_data.columns)
plt.title('Box Plot of Actual Rainfall (2011-2014)')
plt.ylabel('Rainfall (mm)')
plt.grid(True)
plt.show()
```

```python
average_rainfall = rainfall_data.mean()
plt.figure(figsize=(8,5))
average_rainfall.plot(kind='bar', color='green', alpha=0.7)
plt.title('Average Actual Rainfall (2011-2014)')
plt.xlabel('Year')
plt.ylabel('Average Rainfall (mm)')
plt.grid(axis='y')
plt.show()
```

```python
district_name = 'BHOPAL'
district_data = df[df['District'] == district_name][['2011_Actual_Rainfall', '2012_Actual_Rainfall', '2013_Actual_Rainfall', '2014_Actual_Rainfall']].T
district_data.columns = ['Rainfall']
plt.figure(figsize=(8,5))
plt.plot(district_data.index, district_data['Rainfall'], marker='o', linestyle='-', color='red')
plt.title(f'Rainfall Trend in {district_name} (2011-2014)')
plt.xlabel('Year')
plt.ylabel('Rainfall (mm)')
plt.grid(True)
plt.show()

```
